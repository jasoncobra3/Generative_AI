```python
import pandas as pd
messages=pd.read_csv('./SMSSpamCollection/SMSSpamCollection.txt',
                    sep='\t',names=["label","message"])
```

```python
messages
```

```python
## Data cleaning and pre processing
import re
import  nltk
```

```python
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
```

```python
lemmatizer=WordNetLemmatizer()
```

```python
corpus=[]
for i in range (0,len(messages)):
    review=re.sub('[^a-zA-Z]',' ',messages['message'][i])
    review=review.lower()
    review=review.split()
    review=[lemmatizer.lemmatize(word) for word in review if not word in stopwords.words('english')]
    review=' '.join(review)
    corpus.append(review)
    
```

```python
corpus
```

## Create Bag of Words

```python
## create the Bag of Words Model
from sklearn.feature_extraction.text import CountVectorizer
### For Binary BOW enable binary=True ---> Return vector without Frequency of the words
cv=CountVectorizer(max_features=100,binary=False) ## for binary=False
## binary=False means count the number of words (return frequency in Vector)
## max_features defines no. of columns to be selected for creating a vector and select top n features with max frequency for column formaation
```

- max_features=100 basically gives the top 100 words to create a vector / column

```python
cv
```

```python
X=cv.fit_transform(corpus).toarray()
```

```python
X.shape ## 100 columns as we defined max_features=100 in Countvectorizer
```

```python
import numpy as np
np.set_printoptions(edgeitems=30, linewidth=100000, 
    formatter=dict(float=lambda x: "%.3g" % x))
X
```

```python
## For binary=true
cv=CountVectorizer(max_features=100,binary=True) 
```

```python
y=cv.fit_transform(corpus).toarray()
```

```python
import numpy as np
np.set_printoptions(edgeitems=30, linewidth=100000, 
    formatter=dict(float=lambda x: "%.3g" % x))
y
```

##### From above you can see the difference between X and Y Vectors with parameters binary=True and binary=False, X contains the frequency of words in the vectors , while in Y vector it only contains 1 and 0
